###Week 9: Principal Component Analysis (PCA)
Objective: Reduce dimensionality using PCA.

####Experiment 9:

Perform PCA on a high-dimensional dataset and visualize the results.
*   Displays the proportion of variance captured by each principal
component.
*   2D Plot: Projects data onto the first two principal components.
*   3D Plot: Projects data onto the first three principal components.

Note: We can use the Wine Dataset from the sklearn library. It has 13 features (high dimensionality) related to chemical properties of winesamples, which can be reduced to 2 or 3 principal components for visualization.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load the Wine dataset
wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Perform PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# Plot 1: Variance Explained by Principal Components
plt.figure(figsize=(10, 5))
plt.subplot(121)
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio)
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio,
         color='red', linestyle='--', marker='o')
plt.title('Variance Explained by Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.xticks(range(1, len(explained_variance_ratio) + 1))
```

```python
# Annotate individual and cumulative variance
for i, (individual, cumulative) in enumerate(zip(explained_variance_ratio, cumulative_variance_ratio), 1):
    plt.text(i, individual, f'{individual:.3f}', ha='center', va='bottom')
    plt.text(i, cumulative, f'Cumulative: {cumulative:.3f}', ha='center', va='top', color='red')
```

```python
# Plot 2: 2D PCA Visualization
plt.subplot(122)
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title('Wine Dataset - 2D PCA Projection')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.colorbar(scatter, label='Wine Class')

plt.tight_layout()
plt.show()
```

```python
# Plot 3: 3D PCA Visualization
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap='viridis')
ax.set_title('Wine Dataset - 3D PCA Projection')
ax.set_xlabel('First Principal Component')
ax.set_ylabel('Second Principal Component')
ax.set_zlabel('Third Principal Component')
plt.colorbar(scatter, label='Wine Class')

plt.tight_layout()
plt.show()
```

```python
# Print out detailed variance information
print("Variance Explained by Each Principal Component:")
for i, ratio in enumerate(explained_variance_ratio, 1):
    print(f"PC{i}: {ratio:.4f} ({ratio*100:.2f}%)")

print("\nCumulative Variance Explained:")
for i, cumulative in enumerate(np.cumsum(explained_variance_ratio), 1):
    print(f"First {i} PCs: {cumulative:.4f} ({cumulative*100:.2f}%)")
```
